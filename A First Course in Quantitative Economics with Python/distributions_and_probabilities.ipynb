```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import yfinance as yf
import scipy.stats
import seaborn as sns
```

```python
n = 10
u = scipy.stats.randint(1, n+1)
u.mean(), u.var()
```

```python
fig, ax = plt.subplots()

S = np.arange(1,n+1)
ax.plot(S, u.pmf(S), linestyle = "", marker = "o")
ax.vlines(S,0,u.pmf(S)) # plot vertical lines for each x from y min to y max
ax.set_xlabel("S")
ax.set_ylabel("PMF")
ax.set_xticks(S)
plt.show()
```

```python
fig, ax = plt.subplots()
S = np.arange(1,n+1)
ax.step(S,u.cdf(S))
ax.vlines(S,0,u.cdf(S), alpha = 0.2)
ax.set_xticks(S)
ax.set_xlabel("S")
ax.set_ylabel("CDF of a  normal distribution")
plt.show()
```

```python
n = 10
p = 0.35
u = scipy.stats.binom(n,p)

fig, ax = plt.subplots()
S = np.arange(1,n+1)
ax.plot(S,u.pmf(S), marker = "o")
ax.vlines(S,0,u.pmf(S), alpha = 0.2)
ax.set_xticks(S)
ax.set_xlabel("S")
ax.set_ylabel("PMF")
plt.show()
```

```python
μ, σ = 0.0, 1.0
u = scipy.stats.norm(μ, σ)

fig, ax = plt.subplots()
x_grid = np.linspace(-4,4,200)

ax.plot(x_grid, u.pdf(x_grid), label = "Normal PDF")
ax.plot(x_grid, u.cdf(x_grid), label = "Normal CDF")
ax.set_xlabel("x")
ax.set_ylabel("PDF")
plt.legend()
plt.show()
```
