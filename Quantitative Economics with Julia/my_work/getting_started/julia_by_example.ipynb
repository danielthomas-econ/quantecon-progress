```julia
# plotting n draws from some distribution
function genhistogram(distribution,n)
    ep = rand(distribution,n)
    histogram(ep)
end

distr = Poisson() # change this to get a different distribution
genhistogram(distr,10^4)
```

# Exercise 2.1:
Compute $n!$ using a `for` loop

```julia
function factorial2(n)
    result = 1
    for i in 2:n
        result *= i
    end
    return result
end

print(factorial2(20))
```
